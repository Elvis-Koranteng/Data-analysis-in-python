```python
# Data Analysis and Visualization without Seaborn

# Analyze the trends in marijuana arrests over the years using bar plots
df['YEAR'].value_counts().sort_index().plot(kind='bar', figsize=(10, 6))
plt.xlabel('Year')
plt.ylabel('Count')
plt.title('Marijuana Arrests by Year')
plt.show()
```

```python
# Explore relationships between variables using a bar plot
df.groupby('RACE')['AGE'].mean().plot(kind='bar', figsize=(10, 6))
plt.xlabel('Race')
plt.ylabel('Average Age')
plt.title('Average Age of Marijuana Arrests by Race')
plt.show()
```

```python
# Perform aggregations and derive insights using groupby
df.groupby('TYPE')['AGE'].mean().plot(kind='bar', figsize=(10, 6))
plt.xlabel('Type')
plt.ylabel('Average Age')
plt.title('Average Age of Marijuana Arrests by Type')
plt.show()
```

```python
# Visualize the distribution of age across different categories, such as race or type of arrest
df.boxplot(column='AGE', by='RACE', figsize=(10, 6))
plt.xlabel('Race')
plt.ylabel('Age')
plt.title('Distribution of Age by Race')
plt.show()
```

```python
# Compare the counts of arrests by different categories using stacked bar plots
arrests_by_category = df['CATEGORY'].value_counts()
arrests_by_race_gender = df.groupby(['RACE', 'SEX']).size().unstack()

fig, axes = plt.subplots(1, 2, figsize=(12, 6))

arrests_by_category.plot(kind='bar', ax=axes[0])
axes[0].set_xlabel('Category')
axes[0].set_ylabel('Count')
axes[0].set_title('Arrests by Category')

arrests_by_race_gender.plot(kind='bar', stacked=True, ax=axes[1])
axes[1].set_xlabel('Race')
axes[1].set_ylabel('Count')
axes[1].set_title('Arrests by Race and Gender')

plt.tight_layout()
plt.show()
```
